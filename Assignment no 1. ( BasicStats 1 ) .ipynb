### Standardization of Numerical Variables

- scaleing the numerical variables for uniformity and improving the datasetâ€™s suitability for analytical models.

1) Explain the concept of standardization (z-score normalization)?
- Standardization, or z-score normalization, is a way to make data easier to compare. It adjusts the values of a dataset so they have a mean (average)  of 0 and a standard deviation of 1. This helps remove differences in scale, making all features equally important. For example, if one column is in kilograms and another in meters, standardization ensures theyâ€™re on the same scale. Itâ€™s especially useful in machine learning to improve accuracy.

```python
from sklearn.preprocessing import StandardScaler
```

```python
numerical_cols = ['Volume','Avg Price']
Scaler = StandardScaler()
sd[numerical_cols]=Scaler.fit_transform(sd[numerical_cols])

# Displaying a portion of the transformed dataset.

print("Mean after standardization:")
print(sd[numerical_cols].mean())

print("Standard deviation after standardization:")
print(sd[numerical_cols].std())
```

## Conversion of Categorical Data into Dummy Variables

- transforming the categorical variables into a format that can be provided to ML algorithms.

```python
sd.dtypes
```

1) Discuss the need for converting categorical data into dummy variables (one-hot encoding).
- Categorical data, like names or categories, can't be directly used in most machine learning algorithms because they only understand numbers. Converting these categories into dummy variables turns them into a series of 0s and 1s, making the data usable for the algorithms. This ensures each category gets equal importance without implying any order. For example, cities like "Pune" and "Delhi" become separate columns with 0s and 1s. It helps models make better predictions by treating categories correctly.

```python
# Using One Hot Encoding 
# Creating instance variable for all columns
cc = ['Day', 'SKU', 'BU', 'Brand', 'Model']
```

```python
# Apply one-hot encoding
sd_encoded = pd.get_dummies(sd, columns=cc, drop_first=False)
sd_encoded
```

```python
# Ensure binary format (0 and 1)

for col in sd_encoded.columns:
    if sd_encoded[col].dtype == 'bool':  # Check for boolean dtype
        sd_encoded[col] = sd_encoded[col].astype(int)  # Convert to int
```
