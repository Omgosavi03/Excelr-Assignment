# Encoding Techniques

```python
ad.dtypes
```

```python
uv = ad.apply(pd.Series.unique)
uv
```

### Apply One-Hot Encoding to categorical variables with less than 5 categories.

```python
categorical_features = ad.select_dtypes(include=['object']).columns
one_hot_features = [col for col in categorical_features if ad[col].nunique() < 5]
label_features = [col for col in categorical_features if ad[col].nunique() >= 5]
```

```python
# One-Hot Encoding

ad = pd.get_dummies(ad, columns=one_hot_features,drop_first=True)
```

```python
ad
```

### Useing Label Encoding for categorical variables with more than 5 categories.

```python
# Label Encoding
label_encoder = LabelEncoder()
ad = ad.copy()
for col in label_features:
    ad[col] = label_encoder.fit_transform(ad[col])
```

```python
ad
```

1) One-Hot Encoding:
- Pros: It avoids ordinal relationships in categorical data, making it suitable for nominal categories.
- Cons: It increases the dataset size when there are many unique categories.

2) Label Encoding:
- Pros: It is memory-efficient and simple to implement.
- Cons: It can introduce ordinal relationships where they don’t exist, leading to misleading results.

# Feature Engineering:

```python
ad['age_group'] = pd.cut(ad['age'], bins=[0, 30, 50, 100], labels=['Young', 'Middle-aged', 'Senior'])
ad['hours_per_week_log'] = np.log1p(ad['hours_per_week'])
```

# Feature Selection

```python
from sklearn.ensemble import IsolationForest
```

```python
numfeats = ad.select_dtypes(include=['int64', 'float64']).columns
```

```python
# Applying Isolation Forest

isolation_forest = IsolationForest(contamination=0.01, random_state=42)
outliers = isolation_forest.fit_predict(ad[numfeats])
```

```python
ad['outlier'] = outliers
print(ad[['outlier']].value_counts())
```

```python
# Filter out outliers

data_no_outliers = ad[ad['outlier'] == 1]
data_no_outliers
```

```python
# View outliers only

outliers_only = ad[ad['outlier'] == -1]
outliers_only
```

Outliers are extreme values that differ significantly from the rest of the data. They can make a model less reliable by affecting calculations like averages and variances. This can lead to incorrect predictions because the model may try too hard to fit these unusual points. Outliers can also make the model slower to train and less effective on new data, as it becomes less focused on the general patterns in the data. Removing or handling outliers helps improve the model's accuracy and performance.
